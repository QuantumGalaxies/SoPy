```python
import sopy as sp
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
lattices = [np.arange(-10,10.1,0.1)]

v_r = sp.Vector()
for position in [0]:
    v_r += sp.Vector().gaussian(a = 1,positions = [position],sigmas = [1], ls = [0], lattices = lattices)

#v_r = sp.Vector().delta( a = 1 , positions = [0.0], spacings = [1.6] , lattices = lattices , wavenumbers = [5.], phis = [0.])
```

```python
v = sp.Operand( v_r.decompose(1), sp.Vector() ) 
v_r.dot(v_r)
```

```python
plt.plot(lattices[0],v.re.decompose(1)[1][0], label = 'v')
plt.legend()
```

```python
v.exp_i([1.]).cot(v)
#(0.7788007830714049-0.778800783071405j)
```
